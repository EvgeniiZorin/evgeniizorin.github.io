```python
import os

from tqdm import tqdm
from PIL import Image
```

```python
def resize_image(img_path: str, 
                 resize_to: tuple[int, int] = (300, 300)
                 ) -> None:
    # read image
    print(f'Image to resize: {img_path}')
    img = Image.open(img_path)
    before_img_size = img.size
    print(f' > Image size before resizing: {before_img_size}')
    
    # resize
    img = img.resize(resize_to)
    after_img_size = img.size
    print(f' > Image size after resizing: {after_img_size}')
    
    # save resized image
    resized_img_path = os.path.splitext(img_path)[0] + \
        '_RESIZED' + os.path.splitext(img_path)[1]
    img.save(resized_img_path)
    print(f' > Resized image saved to: {resized_img_path}\n')
    return None

# DIRECTORY_WITH_IMAGES = '../images/cpd_books'
DIRECTORY_WITH_IMAGES =   '../images/cpd_courses'
for i in tqdm([j for j in os.listdir(DIRECTORY_WITH_IMAGES) if 'RESIZED' not in j]):
    resize_image(f"{DIRECTORY_WITH_IMAGES}/{i}",
                (300,300)
                )
```
